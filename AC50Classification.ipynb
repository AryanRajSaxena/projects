```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
parameters = {'criterion':('gini', 'entropy'), 'n_estimators':[80,90,100,110,120,130,140,150], 'max_depth': [10,15,20,25,30], 'max_features': ('sqrt', 'log2', None)}
```

```python
cv  = GridSearchCV(RandomForestClassifier(), parameters)
cv.fit(x_train, y_train)

print("Best Hyperparameters:", cv.best_params_)
params = cv.best_params_
```

```python
clf = RandomForestClassifier(n_estimators = params['n_estimators'], criterion = params['criterion'],max_depth = params['max_depth'],max_features = params['max_features'] ,random_state=42)
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)
```

```python
# best_model = clf.best_estimator_
# y_pred = best_model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)

cvc = cross_val_score(clf,x_test,y_test,cv=7)
print(np.mean(cvc))
```
