```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test.csv')
label = pd.read_csv('test_label.csv')
df['label'] = label['0']
```

```python
df
```

```python
label['0'].unique()
```

```python
new_df1 = df[df['label'] == 1]
new_df0 = df[df['label'] == 0]
```

```python
new_df1
```

```python
new_df0 = new_df0[:4600]
```

```python
new_df1.iloc[:,0]
```

```python
fig, axs = plt.subplots(8, 5, figsize=(15, 20))
axs = axs.ravel()

for i in range(38):    
    axs[i].scatter(range(4600), new_df1.iloc[:,i], color='red', s=1)
    axs[i].scatter(range(4600), new_df0.iloc[:,i], color='green', s=1)
    axs[i].set_title(f'Anomaly {i+1}')

plt.tight_layout()
plt.show()
```
